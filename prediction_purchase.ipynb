## Convert the categorical VisitorType feature to numerical using get_dummies() method:

````python
df2['VisitorType'].unique()
````

````python
def convert_categorical_to_numeric(df):
  dummies = pd.get_dummies(df.VisitorType)
  df = pd.concat([df, dummies], axis='columns')
  df = df.drop(columns= 'VisitorType', axis= 'columns')
  return df
````

````python
df2 = convert_categorical_to_numeric(df2)
df3 = convert_categorical_to_numeric(df3)
````

## **__Imbalanced dataset**

````python
X_imbalance = df2.drop("Revenue" , axis = 1)
Y_imbalance = df2.Revenue
````

````python
x_train , x_test , y_train , y_test = train_test_split(X_imbalance,Y_imbalance , test_size = 0.3)
````

````python
x_test.shape
````

````python
x_train.shape
````

````python
knn_imbalance = KNeighborsClassifier(n_neighbors=3)
svc_imbalance = LinearSVC()
LR_imbalance  = LogisticRegression()
````

````python
knn_imbalance.fit(x_train , y_train)
svc_imbalance.fit(x_train , y_train)
LR_imbalance.fit(x_train , y_train)
````

````python
#Evaluation Function

def results(models, X , Y , x_test ,y_test ):
  results = pd.DataFrame(columns =[ 'model_name' , 'f1_score' , 'Precision' , 'Accuracy' ,'Score'])
  for model in models:
    model_name = type(model).__name__
    y_pred = model.predict(X)
    f1 = f1_score(Y , y_pred)
    precision = precision_score(Y, y_pred)
    accuracy = accuracy_score(Y, y_pred)
    Score = model.score(x_test ,y_test)
    row = pd.DataFrame({'model_name': [model_name] ,'f1_score':[f1] , 'Precision': [precision]
                        , 'Accuracy' : [accuracy] ,'Score' : [Score] })
    results = results.append(row)

  return results
````

````python
model_list = [knn_imbalance,svc_imbalance ,LR_imbalance ]
results(model_list ,X_imbalance, Y_imbalance ,x_test ,y_test)
````

**`با توجه به نتایج مشاهده شده مدل بهترین نتیجه را دارد اما با توجه به اینکه داده ها نامتوازن هستند مدل به خوبی کار نکرده است .`**

## **__UnderSampelling**

````python
count_Revenue_0, count_Revenue_1 = df2.Revenue.value_counts()

df_majority = df2[(df2['Revenue']==0)] 
df_minority = df2[(df2['Revenue']==1)] 

df_majority_undersample = df_majority.sample(count_Revenue_1) 
df2_undersample = pd.concat([df_majority_undersample , df_minority ] ,axis = 0)

df2_undersample.Revenue.value_counts()
````

````python
X_undersample = df2_undersample.drop("Revenue",axis=1)
Y_undersample = df2_undersample.Revenue 

x_train,x_test,y_train,y_test = train_test_split(X_undersample,Y_undersample,test_size=0.2)
````

````python
knn_undersample = KNeighborsClassifier(n_neighbors=3)
svc_undersample = LinearSVC(random_state=0, tol=1e-5)
LR_undersample = LogisticRegression()
````

````python
knn_undersample.fit(x_train , y_train)
svc_undersample.fit(x_train , y_train)
LR_undersample.fit(x_train , y_train)
````

````python
models = [knn_undersample , svc_undersample , LR_undersample]
results(models , X_undersample , Y_undersample ,x_test , y_test )
````

**`با توجه به نتایج به دست آمده میبینیم که عملکرد مدل ها بهتر شده است .`**
